## Task 1: NumPy Array Creation

```python
import numpy as np

# 1D Array
arr1 = np.array([7, 12, 14, 19, 21])
print("1D Array:", arr1)

# 2D Array
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print("2D Array:\n", arr2)

# Addition
print("Array + 5:", arr1 + 5)

# Multiplication
print("Array * 2:", arr1 * 2)

# Slicing
print("Sliced Array:", arr1[1:4])
```

## Task 2: Array Statistics

```python
stats_arr = np.array([5, 12, 14, 21, 99])

print("Mean:", np.mean(stats_arr))
print("Median:", np.median(stats_arr))
print("Standard Deviation:", np.std(stats_arr))
print("Max:", np.max(stats_arr))
print("Argmax (Index of Max):", np.argmax(stats_arr))
```

## Task 4: Data Importing  
    We load a CSV file into a Pandas DataFrame and display the first and last few rows.  
    (*Using seabornâ€™s built-in Titanic dataset to avoid downloading files.*)

```python
import seaborn as sns

titanic = sns.load_dataset("titanic")
print("First 5 rows:\n", titanic.head())
print("\nLast 5 rows:\n", titanic.tail())
```

## Task 5: Data Cleaning

```python
print("Missing values before:\n", titanic.isnull().sum())

# Fill missing age with mean
titanic['age'] = titanic['age'].fillna(titanic['age'].mean())

# Fill missing values of deck
titanic['deck'] = titanic['deck'].fillna(titanic['deck'].mode()[0])

# Drop rows where 'embarked' is missing
titanic_cleaned = titanic.dropna(subset=['embarked'])


print("\nMissing values after cleaning:\n", titanic_cleaned.isnull().sum())
```

## Task 6: Indexing & Selection

```python
# Select row with index 0 using loc
print("Using loc:\n", titanic.loc[0])

# Select first row and first 3 columns using iloc
print("\nUsing iloc:\n", titanic.iloc[0, :3])
```

## Task 7: Sorting & Filtering

```python
filtered = titanic[titanic['age'] > 30]
sorted_df = titanic.sort_values(by="fare", ascending=False)

print("Filtered (age > 30):\n", filtered.head())
print("\nSorted by Fare (Top 5):\n", sorted_df.head())
```

## Task 8: GroupBy & Aggregations

```python
grouped = titanic.groupby("class")["age"].mean()
print("Average Age per Class:\n", grouped)
```

## Task 9: Merging & Joining 

```python
df1 = pd.DataFrame({"ID": [1, 2, 3], "Name": ["Ali Zain ul Abedeen", "sawab_e_darain", "MysticDani"]})
df2 = pd.DataFrame({"ID": [1, 2, 3], "Score": [85, 90, 88]})

# Merge
merged = pd.merge(df1, df2, on="ID")
print("Merged DataFrame:\n", merged)

# Concat
concat_df = pd.concat([df1, df2], axis=1)
print("\nConcatenated DataFrame:\n", concat_df)
```

## Task 11: Line Plot Basics

```python
import matplotlib.pyplot as plt

values = [12, 21, 14, 30, 25, 37]
plt.plot(values, marker='o')
plt.title("Line Plot Example")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

## Task 12: Bar Chart Frequencies

```python
titanic['class'].value_counts().plot(kind='bar', color='purple')
plt.title("Passenger Count by Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

## Task 13: Scatter Plot Relationships

```python
plt.scatter(titanic['age'], titanic['fare'], alpha=0.5)
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

## Task 14: Seaborn Pairplot

```python
sns.pairplot(titanic[['age', 'fare', 'survived']], diag_kind='kde')
plt.show()
```

## Task 15: Category Box Plot 

```python
sns.boxplot(x="class", y="age", data=titanic)
plt.title("Age Distribution by Class")
plt.show()
```

## Task 16: Correlation Heatmap

```python
corr = titanic[['age', 'fare', 'survived']].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

## Task 17: Subplot Grid

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line Plot
axs[0, 0].plot(values, marker='o')
axs[0, 0].set_title("Line Plot")

# Bar Chart
titanic['class'].value_counts().plot(kind='bar', ax=axs[0, 1], color='lightgreen')
axs[0, 1].set_title("Class Distribution")

# Scatter Plot
axs[1, 0].scatter(titanic['age'], titanic['fare'], alpha=0.5)
axs[1, 0].set_title("Age vs Fare")

# Histogram
axs[1, 1].hist(titanic['age'].dropna(), bins=20, color='orange')
axs[1, 1].set_title("Age Histogram")

plt.tight_layout()
plt.show()
```

## Task 18: Seaborn Styling

```python
sns.set_palette("Set2")
sns.histplot(titanic['age'], bins=20, kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

## Task 19: Violin Plot Distribution

```python
sns.violinplot(x="class", y="fare", data=titanic)
plt.title("Fare Distribution by Class")
plt.show()
```

## Task 20: Hybrid Visualization

```python
sns.histplot(titanic['age'], bins=30, color="skyblue", stat="density")
sns.kdeplot(titanic['age'], color="red", linewidth=2)
plt.title("Age Distribution with KDE Overlay")
plt.show()
```
